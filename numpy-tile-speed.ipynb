```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
```

```python
def tile_time(A, reps):
    t0 = time()
    y = np.tile(A, reps)
    return time() - t0
```

```python
x = np.random.rand(7)

n = 110


t = np.zeros((n, n))

for i in range(1, n + 1):
    for j in range(1, n + 1):
        t[i - 1][j - 1] = tile_time(x, (i, j))
        
plt.imshow(t, origin='lower')
plt.colorbar()
plt.savefig('/Users/tandav/Desktop/tile-time.png', dpi=300)
```
