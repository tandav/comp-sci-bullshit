# Sphere in parallel coordinates

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
theta = np.linspace(0, 2 * np.pi, n)
phi   = np.linspace(0,     np.pi, n)
r = 1 # radius
```

```python
x = r * np.cos(theta) * np.sin(phi)
y = r * np.sin(theta) * np.sin(phi)
z = r * np.cos(phi)
```

```python
arr = np.vstack((x, y, z)).T
```

```python
arr.shape
```

```python
for i in range(n):
    plt.plot(arr[i])
plt.show()
```
