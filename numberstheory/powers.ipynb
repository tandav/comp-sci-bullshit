# The sum of the digits of n^n https://oeis.org/A066588

```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
matplotlib.style.use('classic')
```

```python
x = range(1000)
y = [i**i for i in x]
```

```python
print(y[:20]) # print first 20 elements
```

```python
z = []
for i in y:
    z.append(sum(int(d) for d in str(i)))
```

```python
print(z[:20]) #  let's see first 20 elements 
```

```python
plt.plot(x, z)
```

```python
plt.show()
```

```python
plt.plot(x[0:100], z[0:100])
```
