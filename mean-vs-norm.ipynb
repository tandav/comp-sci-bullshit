```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 100
means = np.zeros(N)
norms = np.zeros(N)
```

```python
for i in range(N):
    x = np.random.rand(100)
    
    means[i] = np.mean(x)
    norms[i] = np.linalg.norm(x)
```

```python
plt.plot(means * 10) #   blue plot
plt.plot(norms)      # orange plot
```
